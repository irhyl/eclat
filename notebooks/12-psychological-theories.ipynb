## Psychological Theories: Context for éclat

This notebook summarizes psychological theories and empirical findings relevant to conversational financial interfaces: emotion, cognition, decision-making, motivation, stress, and behavior change. Each section explains the theory, why it matters for éclat, and practical design/measurement implications. References are provided for deeper reading.

### 1. Emotion Theories (Basic Emotions, Appraisal, Affective Forecasting)

Classic basic-emotion theories (Ekman) posit discrete emotions with recognizable expressions; appraisal theories (Lazarus, Scherer) emphasize cognitive evaluation shaping emotional responses; affective forecasting describes how people predict future emotional states (often inaccurately).

Relevance to éclat: Real-time emotional signals (sentiment, micro-pauses) can indicate anxiety, confusion, or readiness. Appraisal-focused signals (e.g., perceived fairness, clarity) predict whether a user will continue. Be cautious with affective forecasts — claimed future feelings (e.g., relief on approval) may not match actual outcomes; use them as persuasive aids carefully.

Design implications:
- Detect emotional markers (text sentiment, pause patterns) and adapt tone/modality to reduce anxiety.
- When previewing outcomes (sanction letter), avoid overpromising — present probabilistic language and next steps.

References:
- Ekman, P. (1992). An argument for basic emotions. *Cognition & Emotion*.
- Lazarus, R. S. (1991). Emotion and Adaptation.
- Gilbert, D. T., & Wilson, T. D. (2007). Prospection: Experiencing the future. *Science*.

### 2. Dual-Process & Cognitive Load (System 1 / System 2)

Dual-process models (Kahneman) distinguish fast, intuitive processes (System 1) from slow, deliberative ones (System 2). Cognitive load theory explains how limited working memory affects task performance.

Relevance to éclat: Financial decisions require both quick affordances (clarity, simple choices) and moments for deliberation. Too many simultaneous requests or jargon overloads System 2 and increases abandonment.

Design implications:
- Chunk information and use progressive disclosure; avoid asking for many fields at once.
- Provide short, optional explainer modals for users who switch into deliberative mode.

References:
- Kahneman, D. (2011). *Thinking, Fast and Slow*.
- Sweller, J. (1988). Cognitive load during problem solving. *Cognitive Science*.

### 3. Motivation & Self-Determination Theory (SDT)

SDT (Deci & Ryan) differentiates intrinsic and extrinsic motivation and emphasizes autonomy, competence, and relatedness as core psychological needs.

Relevance to éclat: Applicants are more likely to persist when they feel autonomous (choice), competent (understandable steps), and related (felt connection). Heavy-handed persuasion undermines intrinsic motivation and trust.

Design implications:
- Offer clear choices and simple explanations to support autonomy and competence.
- Use empathic, autonomy-supportive language rather than pressure.

References:
- Deci, E. L., & Ryan, R. M. (2000). The "what" and "why" of goal pursuits: Human needs and the self-determination of behavior. *Psychological Inquiry*.

### 4. Stress, Coping, and Threat Appraisal

Transactional stress theory (Lazarus & Folkman) describes primary and secondary appraisal: whether an event is threatening and whether coping resources are available.

Relevance to éclat: Loan rejections, document requests, or confusing forms trigger threat appraisals. If users perceive low coping resources (complex instructions, slow support), they abandon or escalate.

Design implications:
- Reduce perceived threat by providing stepwise, low-friction remediation options and quick help links.
- Surface small wins (e.g., "Identity verified") to increase perceived coping capacity.

References:
- Lazarus, R. S., & Folkman, S. (1984). *Stress, Appraisal, and Coping*.

### 5. Cognitive Biases & Heuristics (confirmation, loss aversion, anchoring)

People rely on heuristics that introduce predictable biases (Tversky & Kahneman). Common biases include loss aversion, anchoring, confirmation bias, and availability heuristics.

Relevance to éclat: Messaging, ordering of options, and default values can unintentionally anchor user expectations or exploit loss aversion. The system must avoid manipulative defaults and monitor for biased framing.

Design implications:
- Avoid anchoring by offering clear, comparable options and transparency about defaults.
- Monitor outputs (A/B tests) for biased outcomes across demographic segments.

References:
- Tversky, A., & Kahneman, D. (1974). Judgment under uncertainty: Heuristics and biases. *Science*.

### 6. Affective Computing & Emotion Detection

Summary: Affective computing studies automated detection and response to human affect (Picard). Techniques include sentiment analysis, prosody detection, and multimodal fusion.

Relevance to éclat: Automated affect detection can inform pacing, agent empathy, and escalation triggers, but it has accuracy limits and privacy risks.

Design implications:
- Use affect signals as soft triggers (non-blocking) and always allow user override.
- Be transparent about what is detected and why (consent + explainability).

References:
- Picard, R. W. (1997). *Affective Computing*.

### 7. Persuasion, Compliance, and Ethical Nudge Design

Cialdini's principles of persuasion (reciprocity, scarcity, authority, consistency, liking, consensus) show how social influence shapes choices. Ethical nudge design balances behavior change with user autonomy.

Relevance to éclat: The Sales Agent must avoid manipulative persuasion; use ethically constrained nudges that respect autonomy and transparency.

Design implications:
- Implement guardrails (fairness validators) and monitor for undue influence.
- Prefer informational nudges (clarifying benefits/costs) over scarcity/pressure tactics.

References:
- Cialdini, R. B. (2006). *Influence: The Psychology of Persuasion*.
- Thaler, R. H., & Sunstein, C. R. (2008). *Nudge*.

### 8. Habit Formation & Behavior Change Models (COM-B, Fogg)

Summary: COM-B framework (Capability, Opportunity, Motivation → Behavior) and Fogg Behavior Model (motivation, ability, prompt) explain what drives repeated behavior.

Relevance to éclat: Encouraging users to complete applications or remedial steps requires lowering friction (ability), triggering at the right moment (prompt), and supporting motivation.

Design implications:
- Provide timely prompts (e.g., follow-up reminders) tailored to readiness signals.
- Simplify tasks (auto-fill, document upload helpers) to increase ability.

References:
- Michie, S., van Stralen, M. M., & West, R. (2011). The behaviour change wheel: A new method for characterising and designing behaviour change interventions. *Implementation Science*.
- Fogg, B. J. (2009). A behavior model for persuasive design.

### 9. Measurement & Experimental Design Recommendations

Practical suggestions for validating psychological interventions in éclat:

- Use mixed methods: quantitative (drop-off, completion, NPS) and qualitative (user interviews) to measure perceived dignity and trust.
- Pre-register A/B tests when evaluating framing or nudges; monitor subgroup effects to detect disparate impact.
- Use psychometrically validated short scales for anxiety, trust, and perceived fairness in post-task surveys.

References:
- Rubin, D. B. (2008). *Designing Observational Studies* (for causal inference considerations).
- Trochim, W. M. K. (2006). *Research Methods Knowledge Base*.
